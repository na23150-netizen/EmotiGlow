```python
# NOTE:
# Paste this entire script into a Google Colab notebook cell (or multiple cells if you prefer).
# This is a complete, robust, Colab-ready script that:
# 1) Dynamically downloads the FER-2013 CSV data using KaggleHub.
# 2) Parses and preprocesses images (grayscale, 48x48, normalized).
# 3) Filters dataset to 3 emotions: Angry(0), Happy(3), Sad(4).
# 4) Builds a compact CNN using TensorFlow/Keras.
# 5) Trains the model and plots loss/accuracy curves.
# 6) Evaluates on a held-out test set.
# 7) Saves the trained model as emotion_model.h5 and exports TensorFlow Lite:
#    - emotion_model.tflite (float32)
#    - emotion_model_dynamic.tflite (dynamic range quantized)
# 8) Downloads the artifacts (H5, TFLite, labels.json) to your local machine.
#
# Run on a GPU runtime in Colab for best training speed (Runtime -> Change runtime type -> GPU).

# ------------------------------
# 0) Imports, seeds, environment
# ------------------------------
import os
import sys
import json
import time
import zipfile
import random
import shutil
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from sklearn.model_selection import train_test_split

import tensorflow as tf
from tensorflow.keras import layers, models, utils, callbacks
from google.colab import files
import kagglehub # Added for Kaggle dataset download

# Set seeds for reproducibility
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

print("Python:", sys.version.splitlines()[0])
print("TensorFlow:", tf.__version__)
print("GPU available:", tf.config.list_physical_devices('GPU'))

# ------------------------------
# 1) Download fer2013.csv dataset
# ------------------------------
TARGET_CSV = "fer2013.csv"

# Function to find the fer2013.csv file within a base path (case-insensitive)
def find_csv_in_path(base_path, filename=TARGET_CSV):
    target_filename_lower = filename.lower()
    for root, _, files_list in os.walk(base_path):
        for f in files_list:
            if f.lower() == target_filename_lower:
                return os.path.join(root, f)
    return None

# Download FER-2013 dataset using KaggleHub
print("Downloading FER-2013 dataset using KaggleHub...")
kaggle_dataset_path = kagglehub.dataset_download("deadskull7/fer2013")
print("Kaggle dataset downloaded to:", kaggle_dataset_path)

csv_path = find_csv_in_path(kaggle_dataset_path, TARGET_CSV)

# If CSV not found directly, check for zip files and extract them within the Kaggle path
if csv_path is None:
    print(f"{TARGET_CSV} not found directly in {kaggle_dataset_path}. Checking for zip files to extract...")
    zip_files_found = []
    for root, _, files_list in os.walk(kaggle_dataset_path):
        for f in files_list:
            if f.lower().endswith('.zip'):
                zip_files_found.append(os.path.join(root, f))

    if zip_files_found:
        print(f"Found zip files: {zip_files_found}. Attempting to extract...")
        for zip_file_path in zip_files_found:
            try:
                # Extract to the directory where the zip file is located
                extract_dir = os.path.dirname(zip_file_path)
                with zipfile.ZipFile(zip_file_path, 'r') as zf:
                    zf.extractall(extract_dir)
                print(f"Extracted {zip_file_path} to {extract_dir}")
            except Exception as e:
                print(f"Error extracting {zip_file_path}: {e}")

        # After extraction, search again in the original base path (which now contains extracted files)
        csv_path = find_csv_in_path(kaggle_dataset_path, TARGET_CSV)

if csv_path is None:
    raise SystemExit(f"❌ ERROR: {TARGET_CSV} NOT FOUND after KaggleHub download and zip extraction attempts in {kaggle_dataset_path}!")

# Copy the found CSV to the working directory for easier access if it's nested
if not os.path.exists(TARGET_CSV) or os.path.abspath(csv_path) != os.path.abspath(TARGET_CSV):
    shutil.copy(csv_path, TARGET_CSV)
    print(f"Copied '{csv_path}' to '{TARGET_CSV}' in working directory.")
csv_path = TARGET_CSV

print("Using CSV at:", csv_path)

# ------------------------------
# 2) Load CSV and parse pixels
# ------------------------------
print("Loading CSV into pandas (this may take a moment)...")
df = pd.read_csv(csv_path)
print("CSV shape:", df.shape)
required_cols = {'pixels', 'emotion'}
if not required_cols.issubset(set(df.columns)):
    raise SystemExit(f"The CSV does not contain required columns: {required_cols}. Please provide a valid fer2013.csv.")

# Parse pixels column into numpy array using numpy.fromstring for speed
def parse_pixels_series(series):
    N = len(series)
    out = np.empty((N, 48*48), dtype=np.float32)
    for i, s in enumerate(series.values):
        arr = np.fromstring(s, dtype=np.float32, sep=' ')
        if arr.size != 48*48:
            raise ValueError(f"Row {i} in pixels column has {arr.size} values (expected 2304). Check the CSV.")
        out[i] = arr
    return out

print("Parsing pixel strings into numeric arrays...")
pixels_arr = parse_pixels_series(df['pixels'])
X_all = pixels_arr.reshape(-1, 48, 48, 1)
y_all = df['emotion'].values.astype(int)

print("Parsed shapes:", X_all.shape, y_all.shape)
print("Original dataset class distribution:")
print(pd.Series(y_all).value_counts().sort_index())

# ------------------------------
# 3) Filter to Angry(0), Happy(3), Sad(4)
# ------------------------------
ALLOWED = [0, 3, 4]
mask = np.isin(y_all, ALLOWED)
X = X_all[mask]
y = y_all[mask]

# Map labels to 0..2: Angry->0, Happy->1, Sad->2
label_map = {0: 0, 3: 1, 4: 2}
y_mapped = np.vectorize(label_map.get)(y)

print("Filtered dataset shape:", X.shape, y_mapped.shape)
print("Filtered class counts:")
print(pd.Series(y_mapped).value_counts().sort_index())

# Normalize images
X = X.astype('float32') / 255.0
num_classes = 3
y_cat = utils.to_categorical(y_mapped, num_classes)

# ------------------------------
# 4) Train / Validation / Test split
# ------------------------------
# 80% train, 10% val, 10% test
X_temp, X_test, y_temp, y_test = train_test_split(X, y_cat, test_size=0.10, stratify=y_mapped, random_state=SEED)
val_fraction_of_temp = 0.1111111  # ≈0.1 overall
stratify_temp = np.argmax(y_temp, axis=1)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=val_fraction_of_temp,
                                                  stratify=stratify_temp, random_state=SEED)

print("Train/Val/Test shapes:", X_train.shape, X_val.shape, X_test.shape)

# ------------------------------
# 5) Build a compact CNN model
# ------------------------------
def build_cnn(input_shape=(48,48,1), num_classes=3):
    model = models.Sequential(name="EmotionCNN")
    model.add(layers.Conv2D(32, (3,3), activation='relu', padding='same', input_shape=input_shape))
    model.add(layers.BatchNormalization())
    model.add(layers.Conv2D(32, (3,3), activation='relu', padding='same'))
    model.add(layers.MaxPooling2D((2,2)))
    model.add(layers.Dropout(0.25))

    model.add(layers.Conv2D(64, (3,3), activation='relu', padding='same'))
    model.add(layers.BatchNormalization())
    model.add(layers.Conv2D(64, (3,3), activation='relu', padding='same'))
    model.add(layers.MaxPooling2D((2,2)))
    model.add(layers.Dropout(0.25))

    model.add(layers.Conv2D(128, (3,3), activation='relu', padding='same'))
    model.add(layers.MaxPooling2D((2,2)))
    model.add(layers.Dropout(0.3))

    model.add(layers.Flatten())
    model.add(layers.Dense(128, activation='relu'))
    model.add(layers.BatchNormalization())
    model.add(layers.Dropout(0.4))
    model.add(layers.Dense(num_classes, activation='softmax'))
    return model

model = build_cnn()
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

# ------------------------------
# 6) Train the model
# ------------------------------
BATCH_SIZE = 64
EPOCHS = 30
CHECKPOINT_PATH = "best_emotion_model.h5"

es = callbacks.EarlyStopping(monitor='val_loss', patience=6, restore_best_weights=True, verbose=1)
mc = callbacks.ModelCheckpoint(CHECKPOINT_PATH, monitor='val_accuracy', save_best_only=True, verbose=1)
rlr = callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, verbose=1, min_lr=1e-6)

history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    callbacks=[es, mc, rlr],
    verbose=1
)

# ------------------------------
# 7) Plot training loss & accuracy
# ------------------------------
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(history.history.get('loss', []), label='train_loss')
plt.plot(history.history.get('val_loss', []), label='val_loss')
plt.title('Loss')
plt.xlabel('Epoch')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history.get('accuracy', []), label='train_acc')
plt.plot(history.history.get('val_accuracy', []), label='val_acc')
plt.title('Accuracy')
plt.xlabel('Epoch')
plt.legend()
plt.show()

# ------------------------------
# 8) Evaluate on test set
# ------------------------------
if os.path.exists(CHECKPOINT_PATH):
    print("Loading best checkpoint weights for final evaluation...")
    model.load_weights(CHECKPOINT_PATH)

test_loss, test_acc = model.evaluate(X_test, y_test, verbose=2)
print(f"Test loss: {test_loss:.4f}, Test accuracy: {test_acc:.4f}")

# ------------------------------
# 9) Save labels mapping
# ------------------------------
labels = {0: "Angry", 1: "Happy", 2: "Sad"}
with open("labels.json", "w") as f:
    json.dump(labels, f)
print("Saved labels.json")

# ------------------------------
# 10) Save Keras model (.h5)
# ------------------------------
H5_PATH = "emotion_model.h5"
model.save(H5_PATH)
print("Saved Keras model to", H5_PATH)

# ------------------------------
# 11) Convert to TensorFlow Lite
#   - Save float32 TFLite and dynamic-range quantized TFLite
# ------------------------------
TFLITE_FP32 = "emotion_model.tflite"
TFLITE_DYNAMIC = "emotion_model_dynamic.tflite"

print("Converting to TFLite (float32)...")
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()
with open(TFLITE_FP32, "wb") as f:
    f.write(tflite_model)
print("Saved TFLite (float32):", TFLITE_FP32)

print("Converting to TFLite (dynamic range quantization)...")
converter = tf.lite.TFLiteConverter.from_keras_model(model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
tflite_dynamic = converter.convert()
with open(TFLITE_DYNAMIC, "wb") as f:
    f.write(tflite_dynamic)
print("Saved TFLite (dynamic quantized):", TFLITE_DYNAMIC)

# ------------------------------
# 12) Download artifacts to local machine
# ------------------------------
print("Preparing downloads. Colab will prompt you to save these files locally:")
for path in [H5_PATH, TFLITE_FP32, TFLITE_DYNAMIC, "labels.json"]:
    if os.path.exists(path):
        print(" -", path)
        try:
            files.download(path)
        except Exception as e:
            print(f"files.download failed for {path}: {e}")
    else:
        print("Missing:", path)

print("Part 1 complete. You should have downloaded:")
print(" - emotion_model.h5")
print(" - emotion_model.tflite")
print(" - emotion_model_dynamic.tflite")
print(" - labels.json")
```
